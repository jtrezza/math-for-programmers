```python
# The statement asks to draw from -10 to 11, but then the
# solution draws from -10 to 10. So, following the solution:

dots_array = map(lambda x: (x, x**2), range(-10, 11))
# Also possible: [(x,x**2) for x in range(-10,11)]

draw(
    Points(*dots_array),
    grid=(1, 10),
    nice_aspect_ratio=False
)
```

```python
def add(v1, v2):
    return (v1[0] + v2[0], v1[1] + v2[1])
```

```python
dino_vectors2 = [add((-1.5, -2.5), v) for v in dino_vectors]
draw(
    Points(*dino_vectors, color=blue),
    Polygon(*dino_vectors, color=blue),
    Points(*dino_vectors2, color=red),
    Polygon(*dino_vectors2, color=red),
)
```
