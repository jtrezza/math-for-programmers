```python
from math import sqrt
def length(v):
    return sqrt(v[0]**2 + v[1]**2)
```

```python
# The resultant vector and its components
# are scaled equally
print(length((6,4)))
print(length((9,6)))
```

```python
def add(*vectors):
    x = 0
    y = 0
    for a, b in vectors:
        x += a
        y += b
    return (x, y)

add((-2, 0), (1.5, 1.5), (4, 1))
```

```python
def translate(tv, vectors):
    return [add(tv, v) for v in vectors]

translate((1,1), [(0,0), (0,1), (-3,-3)])
```

```python
def scale(s,v):
    return (s*v[0],s*v[1])
```

```python
import math
scale(math.pi, (math.sqrt(2), math.sqrt(3)))
```

```python
def subtract(v1, v2):
    return (v1[0] - v2[0], v1[1] - v2[1])
```

```python
def distance(v1,v2):
    return length(subtract(v1,v2))

def perimeter(vectors):
    distances = [distance(vectors[i], vectors[(i+1)%len(vectors)])
                for i in range(0,len(vectors))]
    return sum(distances)
```

```python
perimeter([(1,0),(1,1),(0,1),(0,0)])
```
